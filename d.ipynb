```python
import pandas as pd
from textblob import TextBlob

df = pd.read_csv('d.csv')

print('complete with prompting')
df_polarity = pd.DataFrame()
df_polarity['prompt'] = df['prompt']

df_subjectivity = pd.DataFrame()
df_subjectivity['prompt'] = df['prompt']

df_length = pd.DataFrame()
df_length['prompt'] = df['prompt']

for col in df.columns[1:]:
    polarities, subjectivities = batch_sentiment(df[col].tolist())
    df_polarity[col] = polarities
    df_subjectivity[col] = subjectivities
    df_length[col] = df[col].apply(len)

display(df)
display(df_polarity)
display(df_subjectivity)
display(df_length)

display(df['prompt'].to_csv('e.csv'))
```

```python
def analyze_sentiment(text):
    """Returns the sentiment, polarity, and subjectivity of thje given text

    Args:
        text (str): a string of which to analyze

    Returns:
        sentiment, polarity, subjectivity: the sentiment (positive, negative, neutral), the polarity score (-1: extremely negative, +1: extremely positive), and subjectivity (0: factual, 1: emotional)
    """
    blob = TextBlob(text)

    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    return polarity, subjectivity

def batch_sentiment(response_list):
    """runs a list of reponses through the analyze_sentiment function

    Args:
        response_list (list): a list of responses, in strings.

    Returns:
        polarity_list, subjectivity_list: lists of the polarity.
    """
    polarity_list, subjectivity_list = [], []
    for response in response_list:
        polarity, subjectivity = analyze_sentiment(response)
        polarity_list.append(polarity)
        subjectivity_list.append(subjectivity)

    return polarity_list, subjectivity_list
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df, df_polarity, df_subjectivity, and df_length are your dataframes

# Function to categorize prompts
def categorize_prompt(prompt):
    if 'boyfreind' in prompt:
        return 'boyfriend'
    elif 'girlfreind' in prompt:
        return 'girlfriend'
    elif 'freind' in prompt:
        return 'friend'
    else:
        return 'other'

# Apply categorization
df['category'] = df['prompt'].apply(categorize_prompt)

# Ensure data is numeric before averaging
df_polarity = df_polarity.apply(pd.to_numeric, errors='coerce')
df_subjectivity = df_subjectivity.apply(pd.to_numeric, errors='coerce')
df_length = df_length.apply(pd.to_numeric, errors='coerce')

# Compute average metrics for each category
avg_polarity = df_polarity.groupby(df['category']).mean()
avg_subjectivity = df_subjectivity.groupby(df['category']).mean()
avg_length = df_length.groupby(df['category']).mean()

# Plotting
sns.set(style="whitegrid")

# Polarity Plot
avg_polarity.plot(kind='bar', figsize=(12, 6))
plt.title('Average Polarity by Prompt Category')
plt.ylabel('Average Polarity')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()

# Subjectivity Plot
avg_subjectivity.plot(kind='bar', figsize=(12, 6))
plt.title('Average Subjectivity by Prompt Category')
plt.ylabel('Average Subjectivity')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()

# Length Plot
avg_length.plot(kind='bar', figsize=(12, 6))
plt.title('Average Length by Prompt Category')
plt.ylabel('Average Length')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```
